## 测度与概率

#### $\S$ 3.2 单调函数与测度的构造

本章从单调函数出发，构造半集代数上的测度，再将此测度扩张到半集代数生成的$\sigma$代数上.

初衷是为解决$\mathscr{B}(\mathbb{R})$上$\sigma$有限测度的存在唯一性问题.

##### 首先给出测度的相关概念
*  给定非空集合$\Omega$中的子集类$C$上的可加测度，有限可加测度，测度(即$\sigma$可加测度)，测度的有限与$\sigma$有限，可测空间，可测集，测度空间，概率(或概率测度)，概率空间.

##### 简单复习增函数的性质
* 2个有趣的例子
* 增函数由一稠集上的值确定，由此可根据需要修改跳跃点处的值

##### 由右连续增函数出发构造出$\mathbb{R}$中左开右闭区间组成的半集代数上的测度.
* 定义非负集函数$\mu((a,b])=F(b)-F(a)$
* 证明$\mu$是$\sigma$可加的，即确为测度，其中一个方向的不等式通过$\mathbb{R}$中有限闭区间的紧性证明

##### 更一般的问题：(测度扩张定理)由半集代数$\ell$上的测度构造由它生成的$\sigma$代数$\sigma(\ell)$上的测度
* 将半集代数$\ell$上的测度扩张到它生成的集代数$\mathscr{A}(\ell)$上
 * 根据引理3.1.6, 对$\forall A\in\mathscr{A}(\ell)$定义$\tilde{\mu}(A)=\sum_{k=1}^{n}\mu(B_{k})$，由$\mu$的$\sigma$可加性及$\tilde{\mu}$的定义可证$\tilde{\mu}$的$\sigma$可加性
* 进而得出半集代数上测度的次可加性和次$\sigma$可加性
 * 证明中用到一种常见的将一组集合变换成两两不交
* 由半集代数$\ell$上的测度$\mu$引出$\Omega$上的外测度
 * 对$\forall A \in \Omega$定义集函数$\mu^{*}(A)\triangleq inf\{\sum_{n=1}^{\infty}\mu(A_{n}): A \subset \bigcup_{n=1}^{\infty}{A_n}, A_n \in \ell, n\in \mathbb{N}\} $
 * 引出的集函数$\mu^{*}$确为外测度，其在$\ell$上的一致性，以及定义中的不降性，次$\sigma$可加性证明都基于下确界定义
* 给出$\mu^{*}$可测集定义，定理3.2.16给出$\Omega$上$\mu^{*}$可测集组成的集合$\mathscr{A}_{\mu}$的重要性质
 * 先根据$\mu^{*}$可测集定义及次可加性证$\mathscr{A}_{\mu}$是集代数，再根据$\mu^{*}$可测集定义证$\mathscr{A}_{\mu}$是$\sigma$代数
* 测度扩张定理，这个扩展就是$\mu$引出的外测度
 * 存在性即证明$\ell$也是$\mu^{*}$可测的
 * 唯一性用到$\lambda$系方法
 * 3.2.19是测度扩张定理在$\mathbb{R}$中左开右闭区间组成的半集代数上的直接推论
