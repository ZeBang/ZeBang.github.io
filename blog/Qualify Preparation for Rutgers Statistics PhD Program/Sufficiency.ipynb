## 充分统计量

 充分统计量不损失信息地把$n$维样本简化为$k$维统计量

* 常见分布族充分统计量的推导


### 1. 常见分布的充分统计量讨论

###### 两点分布$b(1,\theta)$

计算条件分布，$b(1,\theta)$的**可加性**保证了$P(\sum_{i=1}^{n}{X_{i}=t})$的计算，样本$X_1,X_2,...,X_n$的**独立性**保证了$P(X_1=x_1,...,X_{n-1}=x_{n-1},X_{n}=t-\sum_{i=1}^{n-1}{x_{i}})$的计算.

###### 几何分布$Ge(p)$

计算条件分布，$Ge(p)$的**可加性**保证了$P(\sum_{i=1}^{n}{X_{i}=t})$的计算，样本$X_1,X_2,...,X_n$的**独立性**保证了$P(X_1=x_1,...,X_{n-1}=x_{n-1},X_{n}=t-\sum_{i=1}^{n-1}{x_{i}})$的计算.

###### 均匀分布$U(0,\theta)$

因子分解定理，样本联合密度函数为 $p_{\theta}(x)=\theta^{-n}I_{\{0<x_{(n)}<\theta\}}(x)$.

###### 均匀分布$U(\theta_1,\theta_2)$

因子分解定理，样本联合密度函数为 $p_{\theta}(x)={(\theta_2 - \theta_1)}^{-n} I_{\{\theta_1 < x_{(1)} < x_{(n)} < \theta_2 \}}(x)$.

###### 均匀分布$U(\theta,2\theta)$

因子分解定理，样本联合密度函数为 $p_{\theta}(x)={(\theta)}^{-n} I_{\{\theta < x_{(1)} < x_{(n)} < 2\theta \}}(x)$.

###### 泊松分布

法1: 计算条件分布(运用可加性和独立性)

法2: 因子分解定理

法3: 指数族的完备充分统计量为 $T =\sum_{i=1}^{n}{x_{i}}$

###### 负二项分布

因子分解定理

###### $N(\mu, \sigma^2)$分布

因子分解定理，要熟练正态分布中 $\sum_{i=1}^n (x_{i}-\mu)^2$ 的多种形式变换.

###### $LN(\mu, \sigma^2)$分布

因子分解定理

###### $Ga(a,\lambda)$分布

因子分解定理

###### $Be(a,b)$分布

因子分解定理，熟练取对数的技巧 $x^{a}(1-x)^{b}=exp\{a\log(x)+b\log(1-x)\}$

###### 幂分布

因子分解定理

###### 指数分布

法1: 因子分解定理

法2: 计算条件分布，用到 $T =\sum_{i=1}^{n}{x_{i}} \sim Ga(n,\lambda)$

法3: 指数分布的充分完备统计量是 $T =\sum_{i=1}^{n}{x_{i}}$

###### 双参数指数分布

因子分解定理$p(x;\theta,\mu)={\theta}^{-n} e^{\frac{1}{\theta} (n\overline{x}-n\mu)} I_{\{x_{(1)} > \mu \}}(x)$.

### 2. 次序统计量的充分性<font size=0.8>\[1\](p58)</font>

###### 连续分布族下的样本次序统计量是该连续分布族的充分统计量

计算条件分布，$n$个次序统计量组成的统计量 $T=(x_{(1)},x_{(2)},...,x_{(n)})$ 的联合密度函数定理<font size=0.8>\[1\](p43)</font>保证了 $P(X_1=x_1,...,X_{n}=t_{n})$ 的计算.样本 $X_1,X_2,...,X_n$ 的**独立性**保证了 $P(X_1=x_1,...,X_{n}=x_{n})$ 的计算.再根据<font size=0.8>\[1\](p57)</font>引理(其实质就是把连续分布场合下的 $p(X=x,T=t)$ 分离成 $p(X=x)I\{T(X)=t\}$ 就可计算出条件分布.注意要用到 $\prod_{i=1}^{n} p(x_i) I\{T=t\} = \prod_{i=1}^{n} p(x_i=t_i)$

###### 离散分布族下的样本次序统计量是该连续分布族的充分统计量

区别于连续就在于离散场合样本取值可能有某些相同的值.于是算出的条件分布结果，从直观上看，当给定 $T$ 后，样本的可能取值是 $T=(t_1,t_2,..,t_n)$ 的$x$种排列之一. 离散和连续的区别就在于$x$的值不同.
